```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1 /(1 + 1 / np.exp(x))


def relu(x):
    x = x.copy()
    x[x<0] = 0
    return x
```

```python
x = np.linspace(-5, 5, 1000)


fig, ax = plt.subplots(3, sharex=True)
fig.suptitle('Activation Functions')
ax[0].plot(x, sigmoid(x), label="Sigmoid")
ax[1].plot(x, relu(x), label="ReLu")
ax[2].plot(x, np.tanh(x), label="Tanh")
[ax[i].legend(loc="lower right") for i in range(3)]
#plt.savefig("activation_func.pdf")
plt.show()
```
